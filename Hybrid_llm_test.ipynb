```python
import os
from dotenv import load_dotenv
from langchain_ollama import ChatOllama
from langchain_groq import ChatGroq
from langchain_core.prompts import ChatPromptTemplate

# 1. Load environment variables
load_dotenv()

# 2. The Selector Logic
def get_llm():
    llm_type = os.getenv("LLM_TYPE", "ollama")
    
    if llm_type == "groq":
        print(f"ðŸš€ Running on Groq (Office Mode)")
        return ChatGroq(
            model="llama-3.1-8b-instant", # Cloud model
            temperature=0,
            # api_key is auto-loaded from .env
        )
    else:
        print(f"ðŸ  Running on Local Ollama (Home Mode)")
        return ChatOllama(
            model="llama3.1", # Ensure you ran `ollama pull llama3.1` on Mac
            temperature=0
        )

# 3. Initialize the Model
llm = get_llm()

# 4. Create a Prompt Template
# This instructs the AI on HOW to behave.
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a senior software engineer. You are concise and sarcastic."),
    ("human", "{question}")
])

# 5. Build the Chain (The "Runnable")
# Pipe syntax: Prompt -> LLM
chain = prompt | llm

# 6. Run it
question = "Explain what an 'Agent' is in 1 sentence."
print(f"\nUser: {question}")
response = chain.invoke({"question": question})

print(f"AI: {response.content}")
```
